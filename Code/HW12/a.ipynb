```
#There are 6 class labels
#Must implement FSA function with Vapnik Loss (L = sum(u_y - u_k) w/ u = np.dot(X.T,W))

theta = torch.zeros((train_data.shape[1],6), dtype=torch.double)   # Initialize to 36 x 6 matrix
cost = []
lamb = 0.1
eta = 0.01
m  = train_data.shape[0]
p = train_data.shape[1]
mu = 30
k = 30
Niter = 300
for i in range(1,Niter):
  z = torch.matmul(tf_X_train,theta)
  A = (1/(1+(torch.exp(-z))))
  pred = A.argmax(dim=1)
  diff = (pred - train_labels)
  temp = torch.log(1+torch.exp(diff)).sum() + (lamb/(m)) * theta.square_().sum()
  cost.append(temp.detach().numpy())
  temp = temp.clone().detach().requires_grad_(True)
  temp.backward()
  theta = theta - (eta * temp.grad) - (eta * (lamb) * theta)
  temp2 = (Niter-(2*i))/((2*i*mu)+Niter)
  mi = int(k + ((p-k)*max(0,temp2)))
  if mi > k :
    sorted = np.sort(theta.detach().numpy(),axis=0)
    reverse_sorted = np.flip(sorted, axis=None)
    temp1 = reverse_sorted[0:mi,0]
    for j,lol in enumerate(theta):
      if lol in temp1:
        pass
      else:
        theta[j,0]=0
  
```
