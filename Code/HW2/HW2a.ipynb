```
# The analytical solution for linear regression parameters: B = (X^T*X + \lambda*I)^-1 * X^T*Y
# HW specifies \lambda = 0.0001
```

```
Xt = x_train.T

print(x_train.shape, Xt.shape)
```

```
Xp=np.dot(Xt, x_train)
Xp.shape
```

```
I = np.dot(0.0001, np.identity(X.shape[1]))

print(I)
```

```
Yp = np.dot(Xt,y_train)
Yp.shape
```

```
B = np.dot(np.linalg.inv(Xp + I), Yp)

B.shape
```

```
predict_train = np.dot(x_train,B)
```

```
predict_test = np.dot(x_test, B)
```

```
# I develop functions here to iterate this process k times and determine the 
# average r2 score for both the training and test sets on different k random splits

def calculate_estimator(lam, x_train, y_train):
  Xt = x_train.T
  Xp=np.dot(Xt, x_train)
  I = np.dot(lam, np.identity(X.shape[1]))
  Yp = np.dot(Xt,y_train)
  B = np.dot(np.linalg.inv(Xp + I), Yp)
  return B

def aver_r2(k, X, Y, lam):
  r2_train = []
  r2_test = []

  for i in range(k) :
    x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)

    B = calculate_estimator(lam, x_train, y_train)
    predict_train = np.dot(x_train,B)
    predict_test = np.dot(x_test, B)

    Ym = np.mean(y_train)
    y_test_mean = np.mean(y_test)

    r2_train.append(1 - (np.sum(np.power((y_train - predict_train),2)) / np.sum(np.power((y_train - Ym),2))))
    r2_test.append(1 - (np.sum(np.power((y_test - predict_test),2)) / np.sum(np.power((y_test - y_test_mean),2))))

  return r2_train, r2_test, np.mean(r2_train) , np.mean(r2_test)
```

```
train_r2, test_r2, train_averR2 , test_averR2 = aver_r2(10,X,Y,0.0001)
```

```
tab = { 'Train R2 score': train_r2,
       'Test R2 score': test_r2    
}

df = pd.DataFrame(tab)

print(df)

df.to_csv(path + "R2scores.csv",encoding='utf-8')
```
